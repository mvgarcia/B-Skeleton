```python
#Establece una lista con la cantidad de conexiones de cada nodo
def edges(start,stop,out_beta):
    conexion=[]
    for i in range(start,stop):
        cont=0
        for j in range(len(out_beta)):
            if(i==out_beta[j][0]):
                cont=cont+1
        conexion.append(cont)
    return conexion
#retorna la probablidad de conexiones de cada nodo según el número máximo de conecciones 
def probability(edge):
    maximo=np.max(edge)
    ind_prob=[x/maximo for x in edge]
    return ind_prob
#Retorna la cantidad de nodos con número máximo y mínimo de conexiones
def max_min(proba):
    maximum=np.where(proba==np.max(proba))
    l_maximum=len(maximum[0])
    minimum=np.where(proba==np.min(proba))
    l_minimum=len(minimum[0])
    return(np.max(proba),l_maximum,np.min(proba),l_minimum)
```

```python
#Funcion para ver todos los numeros de conexiones
def con(lista):
    new=[] 
    nume=-1
    for j in range(int(max(lista))):      
        nume=nume+1
        for i in range(len(lista)):
            if(lista[i]==nume):
                new.append(nume)
                nume=nume+1
                i=max(lista)
    return new
```

```python
#"numero de puntos con n conexiones" / "el total del numero de pares"
def Nprobabilidad(lista,n):
    contador=0
    for i in range(len(lista)):
        if(lista[i]==n):
             contador+=1
    return contador/len(lista)      
```
