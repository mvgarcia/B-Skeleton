```python
""""
'out_beta' es el archivo del NeighborGraph, 
'complete' es el archivo con todos los datos (el input de NeighborGraph~los datos a procesar)
""""
def graf(out_beta,complete):
    salida_int=[]
    for i in range(len(out_beta)):
        salida_int.append([int(out_beta[i][0]),int(out_beta[i][1])])
    out=np.array(salida_int)
    uno=np.array([complete[out[i][0]] for i in range(len(out))])
    dos=np.array([complete[out[i][1]] for i in range(len(out))])
    a=[]
    for i in range(len(out)):
        a.append(uno[i])
        a.append(dos[i])
    a=np.array(a)
    return(a)
```
